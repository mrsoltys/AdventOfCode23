```python
import pandas as pd
df = pd.read_table("4_data.txt", delimiter="|",names=["Winning","Card"])
df[['Game','WinningNum']] = df.Winning.str.split(":",expand=True) 
df=df.drop(["Winning"],axis=1)
df=df[['Game', 'WinningNum', 'Card']]
df
```

```python
import re
df["WinningNum"]=df.apply(lambda x: re.sub("\s\s+"," ",x.WinningNum).strip(),axis=1)
df["WinningNum"]=df["WinningNum"].str.split(' ')
df["Card"]=df.apply(lambda x: re.sub("\s\s+"," ",x.Card).strip(),axis=1)
df["Card"]=df["Card"].str.split(' ')
df
```

```python
df["Matches"]=0
for i in range(25):
    df["Matches"]=df["Matches"]+df.apply(lambda x: 1 if x.Card[i] in x.WinningNum else 0,axis=1)
df["Score"]=df.apply(lambda x: int(2**(x.Matches-1)),axis=1)
df
```

```python
sum(df["Score"])
```
