There are lots of numbers and symbols you don't really understand, but apparently any number adjacent to a symbol, even diagonally, is a "part number" and should be included in your sum. (Periods (.) do not count as a symbol.)

Here is an example engine schematic:

467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598..

In this schematic, two numbers are not part numbers because they are not adjacent to a symbol: 114 (top right) and 58 (middle right). Every other number is adjacent to a symbol and so is a part number; their sum is 4361. 

```python
import pandas as pd
import re
import numpy as np
text = np.loadtxt("3_data.txt", dtype = str,comments=None)
#Note: Comments=None is key to get # symbols!

```

```python
# need to find all of the special characters
separator=''
specChars=separator.join(text)
specChars=re.sub("[.]","",specChars)
specChars=re.sub("\d","",specChars)
''.join(set(specChars))
```

Ok, I think i'm ready to go. I can use the re.search to find a number, and then use the re.findall to look above, below, and inline for any special characters. If they are found, I'll add that number to the sum. if not, I'll move on.
Either way, i'll need to replace the number with something like !! so re.search can find the next.

```python
partNums=[]
for r in range(len(text)):
    #For each line, search for a number
    match=re.search(r"\d+",text[r])
    while (match != None):
        (start,stop)=match.span()
        num=match.group()
        #Replace number with ! 
        subText="!"*(stop-start)
        text[r]=text[r].replace(num,subText,1) 
        
        #if number is at the beginning of the line, be sure we're not checking -1
        if (start==0):
            start=1
       
        #Check previous line
        if (r!=0):
            test=re.findall("[@#$%&*-+=/-]", text[r-1][start-1:stop+1])
            if (test!=[]):
                partNums.append(int(num))
                match=re.search(r"\d+",text[r])
                continue
                
        #Check current line
        test=re.findall("[@#$%&*-+=/-]", text[r][start-1:stop+1])
        if (test!=[]):
            partNums.append(int(num))
            match=re.search(r"\d+",text[r])
            continue
        
        #Check next line
        if (r!=(len(text)-1)):
            test=re.findall("[@#$%&*-+=/-]", text[r+1][start-1:stop+1])
            if (test!=[]):
                partNums.append(int(num))
                match=re.search(r"\d+",text[r])
                continue
        
        #Search for next number
        print(num)
        match=re.search(r"\d+",text[r])
    
sum(partNums)
```

```python
text = np.loadtxt("3_data.txt", dtype = str,comments=None)
#Note: Comments=None is key to get # symbols!
```

```python
from itertools import chain
gearRatios=[]
for r in range(len(text)):
    #Find Astrisk
    a=re.search("[*]",text[r])
    while (a != None):
        (aStart,aStop)=a.span()
        print(text[r-1][aStart-3:aStop+3])
        print(text[r][aStart-3:aStop+3])
        print(text[r+1][aStart-3:aStop+3])

        #Scan surrounding areas for numbers
        num1=re.finditer(r"\d+",text[r-1][aStart-3:aStop+3])
        num2=re.finditer(r"\d+",text[r][aStart-3:aStop+3])
        num3=re.finditer(r"\d+",text[r+1][aStart-3:aStop+3])

        #Combine list of numbers
        nums=chain(num1,num2,num3)

        #If a number is matching, hold on to it...
        touching=[]
        for match in nums:
            print(match.span(), match.group())
            (start,stop)=match.span()
            if (start==3 or start==4 or stop==3 or stop==4 or (stop-start==3 and start==2)):
                touching.append(int(match.group()))

        print(touching)
        if (len(touching)==2):
            gearRatios.append(touching[0]*touching[1])
            
        #Get rid of astrisk
        text[r]=text[r].replace("*","~",1)
        #Find nextAstrisk
        a=re.search("[*]",text[r])
        
#Check if any of the numbers are touching
#A touching number will have a stopping span of 3 or 4 OR a starting span of 3 or 4
#Or a 3 digit number might have a starting point of 2
```

```python
sum(gearRatios)
```
